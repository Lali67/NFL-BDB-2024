### Prepare data model 1

```python
import pandas as pd
```

```python
file_name_list = ['tracking_week_1.csv', 
                  'tracking_week_2.csv',
                  'tracking_week_3.csv',
                  'tracking_week_4.csv',
                  'tracking_week_5.csv',
                  'tracking_week_6.csv',
                  'tracking_week_7.csv',
                  'tracking_week_8.csv',
                  'tracking_week_9.csv',
                  'plays.csv',
                  'tackles.csv' ]
file_name_list2 = ['tracking_week_1.csv', 
                  'tracking_week_2.csv',
                  'tracking_week_3.csv',
                  'tracking_week_4.csv',
                  'tracking_week_5.csv',
                  'tracking_week_6.csv',
                  'tracking_week_7.csv',
                  'tracking_week_8.csv',
                  'tracking_week_9.csv']
path = './data/'
```

```python
# tracking week data
for item in file_name_list:
    tracking_df_1 = pd.read_csv(path+item)
    tracking_df_1['gamePlayId'] = tracking_df_1['gameId'] * 10000 + tracking_df_1['playId']
    if item != 'plays.csv':
        tracking_df_1['gamePlayNflId'] = tracking_df_1['gameId'] * 1000000000 + \
                                            tracking_df_1['playId']*10000 + tracking_df_1['nflId']
    tracking_df_1.to_csv(path+item, index=False, encoding='utf-8')
    print("Write: ", item)
```

```python
# tracking week data
tracking_df_tackle = pd.DataFrame()
for item in file_name_list2:
    tracking_df_1 = pd.read_csv(path+item)
    if tracking_df_tackle.empty:
        tracking_df_tackle = tracking_df_1#.loc[tracking_df_1['event'] == 'tackle']
    else:
        tracking_df_tackle = pd.concat([tracking_df_tackle, 
                                        tracking_df_1]) #.loc[tracking_df_1['event'] == 'tackle']])
    print("Write: ", item)
tracking_df_tackle.to_csv(path+"tracking_week.csv", index=False, encoding='utf-8')
print("Info -----------------------------")
print(tracking_df_tackle.info(verbose = False))
```

```python
# tracking week data
tracking_df_tackle = pd.DataFrame()
for item in file_name_list2:
    tracking_df_1 = pd.read_csv(path+item)
    if tracking_df_tackle.empty:
        tracking_df_tackle = tracking_df_1.loc[tracking_df_1['event'] == 'tackle']
    else:
        tracking_df_tackle = pd.concat([tracking_df_tackle, 
                                        tracking_df_1.loc[tracking_df_1['event'] == 'tackle']])
    print("Write: ", item)
print("Info -----------------------------")
print(tracking_df_tackle.info(verbose = False))
```
